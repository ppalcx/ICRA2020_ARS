In Robert Bosch Center for Cyber Physical Systems, in the makery lab we perform research on the paradigm of Learning to Control. This essentially means combining the current data driven learning approaches like Deep Learning, Reinforcement Learning with the formal methods present in control theory like PID Control, Hybrid Zero Dynamics etc. In the learning side we heavily use the concept of Reinforcement Learning, a paradigm where an agent learns by interacting with the environment and receiving a scalar signal known as the reward signal or cost function. 

To perform learning we extensively use the PyBullet Simulator. To perform reinforcement learning we need to create an environment. Below is the example of our environment

```python
import sys, os
sys.path.append(os.path.realpath('../..'))
import pybullet as p
import numpy as np
import time
import pybRL.envs.stoch2_gym_bullet_env_bezier as e
PI = np.pi

walk = [0, PI, PI/2, 3*PI/2]
pace = [0, PI, 0, PI]
bound = [0, 0, PI, PI]
trot = [0, PI, PI , 0]
custom_phase = [0, PI, PI+0.1 , 0.1]

env = e.Stoch2Env(render = True, phase = trot, stairs = False, on_rack=False, gait = "trot")
```

Every environment possesses a state, an action and a reward. In our environment the state is the orientation of the robot, the action is the set of 18 spline points that describe the foot trajectory and the reward is the set of distance moved - Energy expenditure

```python
for i in range(10):
    state, reward, done, info = env.step(np.ones(18))
    print("state: "+ str(state)+" reward: "+ str(reward))
```
